```python
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.python.platform import gfile
import os.path
import matplotlib.pyplot as plt

file_list=[]
file_glob = os.path.join('/home/johan/repos/GitHub/Deep_Learning_UAO/image_aug', '*.jpg')
file_list.extend(gfile.Glob(file_glob))
c = 1
```

```python
for image in file_list:
    img = cv2.imread(image)
    
    #rotacion 90 grados
    rows,cols,depth = img.shape        
    M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
    dst = cv2.warpAffine(img,M,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dst)
    c = c+1
    #rotacion 45 grados           
    N = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
    dstN = cv2.warpAffine(img,N,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstN)
    c = c+1
    #rotacion a 180 grados
    P = cv2.getRotationMatrix2D((cols/2,rows/2),180,1)
    dstP = cv2.warpAffine(img,P,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstP)
    c = c+1
    #rotacion a 25 grados
    Q = cv2.getRotationMatrix2D((cols/2,rows/2),25,1)
    dstQ= cv2.warpAffine(img,Q,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstQ)
    c = c+1
    #rotacion a 270 grados
    R = cv2.getRotationMatrix2D((cols/2,rows/2),270,1)
    dstR= cv2.warpAffine(img,R,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstR)
    c = c+1
    #rotacion 120 grados escalado 2
    rows,cols,depth = img.shape        
    D = cv2.getRotationMatrix2D((cols/2,rows/2),120,2)
    dstD = cv2.warpAffine(img,D,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstD)
    c = c+1
    #rotacion 120 grados escalado 0.5
    rows,cols,depth = img.shape        
    A = cv2.getRotationMatrix2D((cols/2,rows/2),130,0.5)
    dstA = cv2.warpAffine(img,A,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstA)
    c = c+1
    #rotacion 75 grados  escalado 2         
    C = cv2.getRotationMatrix2D((cols/2,rows/2),75,2)
    dstC = cv2.warpAffine(img,C,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstC)
    c = c+1
    #rotacion a 180 grados escalado 2
    B = cv2.getRotationMatrix2D((cols/2,rows/2),210,2)
    dstB = cv2.warpAffine(img,B,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstB)
    c = c+1
    #rotacion a 55 grados escalado 2
    D = cv2.getRotationMatrix2D((cols/2,rows/2),55,2)
    dstD = cv2.warpAffine(img,D,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstD)
    c = c+1
    #rotacion a 300 grados escalado 2
    E = cv2.getRotationMatrix2D((cols/2,rows/2),300,2)
    dstE= cv2.warpAffine(img,E,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstE)
    c = c+1
    #rotacion 75 grados  escalado 0.5         
    C = cv2.getRotationMatrix2D((cols/2,rows/2),75,0.5)
    dstC = cv2.warpAffine(img,C,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstC)
    c = c+1
    #rotacion a 180 grados escalado 0.5
    B = cv2.getRotationMatrix2D((cols/2,rows/2),210,0.5)
    dstB = cv2.warpAffine(img,B,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstB)
    c = c+1
    #rotacion a 55 grados escalado 0.5
    D = cv2.getRotationMatrix2D((cols/2,rows/2),55,0.5)
    dstD = cv2.warpAffine(img,D,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstD)
    c = c+1
    #rotacion a 300 grados escalado 0.5
    E = cv2.getRotationMatrix2D((cols/2,rows/2),300,0.5)
    dstE= cv2.warpAffine(img,E,(cols,rows))
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',dstE)
    c = c+1
    #borroneo gaussiano original
    blur = cv2.GaussianBlur(img,(9,9),0)
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',blur)
    c = c+1
    #borroneo gaussiano rotada 55 grados escalado 0.5
    blur = cv2.GaussianBlur(dstD,(9,9),0)
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',blur)
    c = c+1
    #ruido gaussiano distirbucion normal
    noise = np.zeros(img.shape)
    m = (0,0,0)
    s = (99,99,99)    
    cv2.randn(noise,m,s)
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',noise+img)
    c = c+1
    #ruido gaussiano distirbucion uniforme
    noise1 = np.zeros(img.shape)       
    cv2.randu(noise1,m,s)
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',noise1+img)
    c = c+1
    #laplaciano
    laplacian = cv2.Laplacian(img,cv2.CV_64F)
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',laplacian+img)
    c = c+1
    #flip vertical
    flipup = cv2.flip(img,1)
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',flipup)
    c = c+1
    #flip horizontal
    fliphor = cv2.flip(img,0)
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',fliphor)
    c = c+1
    #flip both
    flipboth = cv2.flip(img,-1)
    cv2.imwrite('images_aug\image_'+ str(c) + '.jpg',flipboth)
    c = c+1
```
